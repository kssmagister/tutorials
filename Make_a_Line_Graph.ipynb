# Make a Line Graph

```
import pandas as pd
y = {'f(x)': [2,10,6,8], 'g(x)': [3,-1,3,5]}
x = [1,2,3,4]

graph = pd.DataFrame(y,x)
graph.plot(kind='line', grid=True, title="Meine Graphik", ylabel="y-Achse", xlabel="x-Achse")
```
